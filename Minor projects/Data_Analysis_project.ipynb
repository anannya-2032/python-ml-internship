```python
import pandas as pd
import numpy as np
import seaborn as sns

pstore=pd.read_csv(r"C:\Users\Dell\Desktop\PYTHON ML INTERNSHIP\database\googleplaystore.csv")
pstore.head(50)
```

```python
pstore.shape
```

```python
print(pstore.info())
print(pstore.describe(include="all"))
```

```python
print(pstore.isnull().sum())
```

```python
pstore=pstore.dropna()
print(pstore.isnull().sum())
```

```python
pstore["Reviews"]=pd.to_numeric(pstore["Reviews"],errors="coerce")
pstore
```

```python
pstore["Installs"]=pstore["Installs"].str.replace("+","").str.replace(",","").astype(int)
```

```python
pstore
```

```python
pstore.info()
```

```python
pstore["Price"]=pstore["Price"].str.replace('$', '').astype(float)
pstore
```

```python
pstore['LastUpdated'] = pd.to_datetime(pstore['LastUpdated'])
pstore
```

```python
pstore = pstore.drop_duplicates()
pstore.shape
```

```python
pstore["Size"]=pstore["Size"].replace("Varies with device",np.nan)
pstore
```

```python
def convert_size(size):
    if "M" in size:
        return float(size.replace("M"," "))
    elif "k" in size:
        return float(size.replace("k"," "))/1000
    else:
        return np.nan

pstore["Size"]=pstore["Size"].apply(lambda x:convert_size(x) if isinstance (x,str) else x)
pstore
```

```python
category_data = pstore.groupby('Category')[['Rating', 'Reviews',"Installs"]].mean().round(2)
print(category_data)
```

```python
correlation=pstore[["Reviews","Rating","Size","Installs","Price"]].corr()
correlation
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True , cmap="crest")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x="Category",y="Installs",data=pstore,estimator=np.mean, errorbar=("ci",100))
plt.xticks(rotation=90)
plt.title("Average Installs by Category", fontsize=18)
plt.show()
```

```python
avg_installs_by_type=pstore.groupby(["Category","Type"])["Installs"].mean().reset_index()

plt.figure(figsize=(14,8))
sns.barplot(x="Category",y="Installs",hue="Type",data=avg_installs_by_type, palette="pastel")
plt.xticks(rotation=90)
plt.title("Average Number of Installs By Category (Free vs Paid)",fontsize=20)
plt.xlabel("Average Number of Installs",fontsize=15)
plt.ylabel("Category",fontsize=15)
plt.legend(title="Type",fontsize=15)
plt.show()
```
