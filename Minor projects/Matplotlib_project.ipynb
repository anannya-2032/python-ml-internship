```python
import seaborn as sns
import pandas as pd
import numpy as np

df=pd.read_csv(r"C:\Users\Dell\Desktop\PYTHON ML INTERNSHIP\database\googleplaystore.csv")
df.head(50)
```

```python
df.shape
```

```python
print(df.info())
print(df.describe(include="all"))
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna()
```

```python
print(df.isnull().sum())
```

```python
df["Reviews"]=pd.to_numeric(df["Reviews"],errors="coerce")
```

```python
df
```

```python
df["Installs"]=df["Installs"].str.replace("+", "").str.replace(",","")
df
```

```python
df["Price"]=df["Price"].str.replace('$', '').astype(float)
df
```

```python
df['LastUpdated'] = pd.to_datetime(df['LastUpdated'])
df
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df["Size"]=df["Size"].replace("Varies with device",np.nan)
df
```

```python
def convert_size(size):
    if "M" in size:
        return float(size.replace("M"," "))
    elif "k" in size:
        return float(size.replace("k"," "))/1000
    else:
        return np.nan

df["Size"]=df["Size"].apply(lambda x:convert_size(x) if isinstance (x,str) else x)
df
```

```python
category_data = df.groupby('Category')[['Rating', 'Reviews',"Installs"]].mean().round(2)
print(category_data)
```

```python
correlation=df[["Reviews","Rating","Size","Price"]].corr()
correlation
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True , cmap="crest")
plt.title("Correlation Heatmap")
plt.show()
            
```
